### Relation Target/Features 

#### Sous ensembles positifs négatifs 

```python
positive_df = df[df['SARS-Cov-2 exam result'] == 'positive']
```

```python
negative_df = df[df['SARS-Cov-2 exam result'] == 'negative']
```

#### Sous ensembles Blood & Viral 

```python
missing_rate = df.isna().sum()/df.shape[0]
```

```python
blood_columns = df.columns[(missing_rate < 0.9) & (missing_rate > 0.88)]
```

```python
viral_columns = df.columns[(missing_rate < 0.88) & (missing_rate > 0.75)]
```

#### Target/Blood relation 

```python
for col in blood_columns: 
    plt.figure()
    sns.distplot(positive_df[col].dropna(), label="positive")
    sns.distplot(negative_df[col].dropna(), label="negative")
    plt.legend()
```

L'objectif est de voir s'il y a une différence entre les cas positifs et négatifs. 
Il semblerait que les gens positifs au Covid ont un taux de platelets différents de ceux quin sont négatifs au covid -> Il faut tester cette hypothèse

#### Création d'une nouvelle variable 'is_sick'

```python
df['is_sick'] = np.sum(df[viral_columns[:-2]] == 'detected', axis=1) >= 1
```

```python
#### Est ce qu'il y a une relation entre les variables blood et le fait d'être malade
malade_df = df[df['is_sick'] == True]
non_malade_df = df[df['is_sick'] == False]

for col in blood_columns:
    plt.figure()
    sns.distplot(malade_df[col].dropna())
    sns.distplot(non_malade_df[col].dropna())
```

Apparemment, les lymphocytes des patients malades et non malades sont différents. (à tester)

### Tests Statistiques 

```python
# Lors d'un test de student il est préferable que les classes aient la même taille
positive = positive_df
negative = negative_df.sample(positive_df.shape[0])
```

```python
# test de student  
from scipy.stats import ttest_ind
def t_test(col,alpha):
    positive = positive_df
    negative = negative_df.sample(positive_df.shape[0])
    
    stat, p = ttest_ind(negative[col].dropna(),positive[col].dropna())
    if (p < alpha):
        return 'Rejet de H0'
    else: 
        return 'Accepter H0'
```
